<h3>IMPORTING...</h3>

```python
# import necessary libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
sns.set(rc={'figure.figsize':(11.7,8.27)})

from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
```

<h3>WRANGLING...</h3>

```python
# import file
path = './ultimate_data_challenge.json'
df = pd.read_json(path)
df.sample(10)
```

```python
# info
df.info()
```

```python
# converting to datetime
def convert_to_datetime(df, col_names):
    for i in col_names:
        df[i] = pd.to_datetime(df[i])
    
    df.info()
    return df
        
df  = convert_to_datetime(df, ['signup_date', 'last_trip_date'])
```

```python
# sorting
df = df.sort_values(by='last_trip_date')
df
```

```python
# checking unique values
df['city'].unique()
```

<p><b>NOTE: </b>we could tackle this problem for each city separately (by unique characteristics), but we will stick as general.</p>

```python
# adding target column
df['target'] = df['last_trip_date'].apply(lambda x: 0 if x.month < 6 else 1)
df
```

```python
# dropping unrelevant columns
df.columns
df = df[['target', 'trips_in_first_30_days', 'avg_rating_of_driver', 'avg_rating_by_driver',
       'avg_surge', 'avg_dist', 'phone', 'surge_pct',
       'ultimate_black_user', 'weekday_pct'
        ]]
df
```

```python
# dropping NaNs from 'phone' column because it's a small portion and excluding making us more bias. Also making 'phone' numeric.
df = df.dropna(subset=['phone'])
df['phone_int'] = df['phone'].apply(lambda x: 0 if x == 'iPhone' else 1)
df['phone_int'].unique()
```

```python
# check how many NaNs in each column
for i in list(df.columns):
    print(f"{i} : {df[i].isna().sum()}")
```

```python
# looking ad destribution
for i in ['avg_rating_of_driver', 'avg_rating_by_driver', 'phone_int']:
    sns.distplot(df[i]).set_title(i)
    plt.show()
```

```python
# because columns 'avg_rating_of_driver' and 'avg_rating_by_driver' skwed to the left we will replace NaNs with 'median' (median values).
for i in ['avg_rating_of_driver', 'avg_rating_by_driver']:
    df[i] = df[i].fillna(df[i].median())
    
for i in list(df.columns):
    print(f"{i} : {df[i].isna().sum()}")
```

```python
# making 'ultimate_black_user' numeric.
df['ultimate_black_user_int'] = df['ultimate_black_user'].apply(lambda x: 0 if x == False else 1)
df['ultimate_black_user_int'].unique()
```

<h3>EDA...</h3>

```python
# dropping categorical columns (not numerical)
df = df[['target', 'trips_in_first_30_days', 'avg_rating_of_driver','avg_rating_by_driver', 'avg_surge', 
         'avg_dist', 'surge_pct', 'weekday_pct', 'phone_int','ultimate_black_user_int']]
df.head(3)
```

```python
print(df.corr())
```

```python
# creating hitmap
sns.heatmap(df.corr(), annot=True)
```

<p><b>NOTE: </b>As we can see, the most correlated features are 'trips_in_first_30_days', 'phone_int', and 'ultimate_black_user_int'.</p>

<h3>MODELING...</h3>
<p>It's a classification problem. So, we are going to use SVM as our prime learning algorithm. </p>

```python
# splitting data
X = df[['trips_in_first_30_days', 'avg_rating_of_driver',
       'avg_rating_by_driver', 'avg_surge', 'avg_dist', 'surge_pct',
       'weekday_pct', 'phone_int', 'ultimate_black_user_int']]
y = df['target']

X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# train the model using training sets
model_svc = svm.SVC()
model_svc.fit(X_train, y_train)
```

```python
# predict the response for the dataset
y_pred = model_svc.predict(X_test)
```

```python
# model accuracy
print(f"Accuracy: {metrics.accuracy_score(y_test, y_pred)}")
```

```python
# model precision
print(f"Precision: {metrics.precision_score(y_test, y_pred)}")
```

```python
# model recall
print(f"Recall: {metrics.recall_score(y_test, y_pred)}")
```

```python
# confusion matrix
confusion_matrix(y_test, y_pred)
```

```python
# train the model using training sets
model_svc = svm.SVC()
model_rfc = RandomForestClassifier()
model_nb = MultinomialNB()

model_list = [model_svc, model_rfc, model_nb]
for model in model_list:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"Accuracy: {metrics.accuracy_score(y_test, y_pred)}")
    print(f"Precision: {metrics.precision_score(y_test, y_pred)}")
    print(f"Recall: {metrics.recall_score(y_test, y_pred)}")
    print(f"Confusion matrix: {confusion_matrix(y_test, y_pred)}")
    print("\n")
```

```python
# let's stick with random forest tree and try tunning it
n_estimators = [x for x in range(150, 1000, 10)]
max_features = ['auto', 'sqrt']
max_depth = [x for x in range(10, 110, 11)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]
```

```python
# making random grid
random_grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'max_depth': max_depth,
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf,
    'bootstrap': bootstrap,
}
```

```python
# checking random grid
print(random_grid)
```

```python
# making model
model_rf_tun = RandomForestClassifier()
rf_random = RandomizedSearchCV(estimator=model_rf_tun, param_distributions=random_grid,
                               n_iter=100, cv=3, verbose=2, random_state=3, n_jobs=-1)
```

```python
# fitting the random search model
rf_random.fit(X_train, y_train)
```

```python
# best parameters from fitting the random search
rf_random.best_params_
```

```python
# looking at prediction of the model with best hyperparemeters
best_model = rf_random.best_estimator_
y_pred = best_model.predict(X_test)
print(f"Accuracy: {metrics.accuracy_score(y_test, y_pred)}")
print(f"Precision: {metrics.precision_score(y_test, y_pred)}")
print(f"Recall: {metrics.recall_score(y_test, y_pred)}")
print(f"Confusion matrix: {confusion_matrix(y_test, y_pred)}")
print("\n")
```
